*   **Pregunta 1.**

Busca y elimina todos los saltos de línea '\n' que se encuentran al final de cada comentario.

Una vez finalizado, imprime los primeros 10 comentarios del resultado obtenido.


```python
docs = [doc.rstrip('\n') for doc in docs]
```

*   **Pregunta 2.**  

Busca e imprime todas las palabras que terminan con dos o más signos de admiración seguidos, por ejemplo "!!!".

Debes imprimir tanto la palabra como la totalidad de signos de admiración que le siguen.

Indica cuántos resultados obtuviste.



```python
exlamacion_lst = []
for comentario in docs:
    palabras = comentario.split()
    for palabra in palabras:
        matches = re.findall(r'\w*!{2,}', palabra)
        for match in matches:
            exlamacion_lst.append(match)
            print(f'Palabra: {match}')
```

```python
print(f'Total de palabras con signos de exclamacion {len(exlamacion_lst)}')
```

*   **Pregunta 3.**  

Busca e imprime todas las palabras que están escritas totalmente en mayúsculas. Cada coincidencia debe ser una sola palabra.

Indica cuántas palabras encontraste.



```python
mayuscula_lst = []
for comentario in docs:
    palabras = comentario.split()
    for palabra in palabras:
        if palabra.isupper():
            mayuscula_lst.append(palabra)
            print("Palabra en mayúsculas:", palabra)
```

```python
print(f'Total de palabras en mayuscula {len(mayuscula_lst)}')
```

*   **Pregunta 4.**  

Busca e imprime los comentarios en donde todos los caracteres alfabéticos (letras) están en mayúsculas.

Cada coincidencia encontrada debe ser todo el comentario/enunciado.

Indica cuántos resultados obtuviste.


```python
comentario_mayus_lst = []
for comentario in docs:
    if comentario.strip().isupper():  # Verificar si todos los caracteres alfabéticos están en mayúsculas
        comentario_mayus_lst.append(comentario)
        print("Comentario en mayúsculas:", comentario.strip())
```

```python
print(f'Total de comentarios en mayuscula {len(comentario_mayus_lst)}')
```

*   **Pregunta 5.**  

Busca e imprime todas las palabras que tengan una vocal acentuada, del tipo á, é, í, ó, ú.

Indica cuántos resultados obtuviste.

```python

palabras_con_vocal_acentuada = []

for comentario in docs:
    palabras = comentario.split()
    for palabra in palabras:
        if re.search(r'[áéíóú]', palabra, re.IGNORECASE):
            palabras_con_vocal_acentuada.append(palabra)

for palabra in palabras_con_vocal_acentuada:
    print("Palabra con vocal acentuada:", palabra)
```

```python
print("Número de resultados:", len(palabras_con_vocal_acentuada))
```

*   **Pregunta 6.**  

Busca e imprime todas las cantidades numéricas monetarias, enteras o con decimales, que inician con el símbolo $\$$.

Indica cuántos resultados obtuviste.

```python
cantidades_monetarias = []

for comentario in docs:
    cantidades = re.findall(r'\$\d+(?:\.\d+)?', comentario)
    cantidades_monetarias.extend(cantidades)

for cantidad in cantidades_monetarias:
    print("Cantidad monetaria:", cantidad)
```

```python
print("Número de resultados:", len(cantidades_monetarias))
```

*   **Pregunta 7.**  

Busca e imprime todas las palabras que sean variantes de la palabra "love", sin importar si incluyen mayúsculas o minúsculas, o la manera en que esté conjugada o alguna otra variación que se haga con dicha palabra.

Indica cuántos resultados obtuviste.

```python
variantes_love = []

for comentario in docs:
    palabras = comentario.split()
    for palabra in palabras:
        if re.search(r'love', palabra, re.IGNORECASE):
            variantes_love.append(palabra)

for variante in variantes_love:
    print("Variante de 'love':", variante)
```

```python
print("Número de resultados:", len(variantes_love))
```

*   **Pregunta 8.**  

Busca e imprime todas las palabras, variantes de "so" y "good", que tengan dos o más "o" en "so" y 3 o más "o" en good.

Indica cuántas encontraste.


```python
palabras_bien = []

for comentario in docs:
    palabras = comentario.split()
    for palabra in palabras:
        if re.search(r'so{2,}', palabra, re.IGNORECASE) or re.search(r'goo{3,}d', palabra, re.IGNORECASE):
            palabras_bien.append(palabra)

for palabra in palabras_bien:
    print("Palabra encontrada:", palabra)

```

```python
print("Número de resultados encontrados:", len(palabras_bien))
```

*   **Pregunta 9.**  

Busca e imprime todas las palabras que tengan una longitud mayor estrictamente a 10 caracteres alfabéticos.

No se consideran los signos de puntuación o caracteres especiales en la longitud de estas cadenas, solo caracteres alfabéticos en mayúsculas o minúsculas.

Indica la cantidad de palabras encontradas.


```python
palabras_len = []

for comentario in docs:
    palabras = re.findall(r'\b[a-zA-Z]+\b', comentario)  # Extraer palabras alfabéticas
    for palabra in palabras:
        if len(palabra) > 10:
            palabras_len.append(palabra)

for palabra in palabras_len:
    print("Palabra encontrada:", palabra)
```

```python
print("Cantidad de palabras encontradas:", len(palabras_len))
```

*   **Pregunta 10.**  

Busca e imprime todas las palabras que inician con una letra mayúscula y terminan con una minúscula, pero que además no sea la primera palabra del comentario/string.

Indica la cantidad de resultados obtenidos.

```python
palabras_mayus_minus = []

for comentario in docs:
    palabras = re.findall(r'\b[A-Z][a-z]+[a-z]\b', comentario)  # Buscar palabras que inician con mayúscula y terminan con minúscula
    if len(palabras) > 1:  # Verificar si hay más de una palabra en el comentario
        for palabra in palabras[1:]:  # Ignorar la primera palabra
            palabras_mayus_minus.append(palabra)

for palabra in palabras_mayus_minus:
    print("Palabra encontrada:", palabra)
```

```python
print("Cantidad de palabras encontradas:", len(palabras_mayus_minus))
```

*   **Pregunta 11.**  

Busca e imprime la secuencia de dos o más palabras que están separadas por un guion, "-", sin que tengan espacios en blanco entre ellas.

Por ejemplo "Go-Kart" sería válido, pero "Go  -Kart" o "Go  -  Kart" no lo serían.

Indica la cantidad de resultados obtenidos.

```python
secuencias_encontradas = []

for comentario in docs:
    secuencias = re.findall(r'\b[A-Za-z]+-[A-Za-z]+\b', comentario)  # Buscar secuencias de palabras separadas por un guion
    secuencias_encontradas.extend(secuencias)

for secuencia in secuencias_encontradas:
    print("Secuencia encontrada:", secuencia)
```

```python
print("Cantidad de secuencias encontradas:", len(secuencias_encontradas))
```

*   **Pregunta 12.**  

Busca e imprime todas las palabras que terminan en "ing" o "ed".

Indica la cantidad de palabras que encontraste de cada una.

```python
palabras_ing = []
palabras_ed = []

for comentario in docs:
    palabras = re.findall(r'\b\w+ing\b', comentario)  # Buscar palabras que terminan en "ing"
    palabras_ing.extend(palabras)
    palabras = re.findall(r'\b\w+ed\b', comentario)  # Buscar palabras que terminan en "ed"
    palabras_ed.extend(palabras)

for palabra in palabras_ing:
    print("Palabra que termina en 'ing':", palabra)
for palabra in palabras_ed:
    print("Palabra que termina en 'ed':", palabra)
```

```python

print("Cantidad de palabras que terminan en 'ing':", len(palabras_ing))
print("Cantidad de palabras que terminan en 'ed':", len(palabras_ed))
```

#**Parte 3. Proceso de limpieza.**

*   **Pregunta 13.**  

Ahora realiza un proceso de limpieza del corpus que incluya los siguientes procesos:

*   Solo se deben considerar caracteres alfabéticos. Es decir, se eliminan todos los signos de puntuación y caracteres especiales.
*   Todos los caracteres alfabéticos se transforman a minúsculas.
*   Se deben eliminar todos los espacios en blanco adicionales que se puedan encontrar en cada comentario.

Al finalizar dicho proceso de limpieza, imprime el resultado de los primeros 10 comentarios resultantes.
   




```python
comentarios_limpios = []

for comentario in docs:
    # Eliminar signos de puntuación y caracteres especiales, transformar a minúsculas y eliminar espacios en blanco adicionales
    comentario_limpio = re.sub(r'[^a-zA-Z\s]', '', comentario).lower().strip()
    # Eliminar espacios en blanco adicionales
    comentario_limpio = re.sub(r'\s+', ' ', comentario_limpio)
    comentarios_limpios.append(comentario_limpio)
for i in range(10):
    print(comentarios_limpios[i])
```

*   **Pregunta 14.**  

Con el resultado de la limpieza obtenido en la pregunta anterior, realiza ahora un proceso de tokenización por palabras del corpus.

Es decir, al final de este proceso de tokenización, debes tener como resultado una lista de listas, donde cada comentario estará tokenizado por palabras.

Al terminar calcula el total de tokens obtenido en todo el corpus.

```python
# Proceso de tokenización
corpus_tokenizado = [comentario.split() for comentario in comentarios_limpios]

# Calcular el total de tokens en todo el corpus
total_tokens = sum(len(comentario) for comentario in corpus_tokenizado)

# Imprimir el corpus tokenizado y el total de tokens
print("Corpus tokenizado:")
for comentario in corpus_tokenizado:
    print(comentario)
print("\nTotal de tokens en todo el corpus:", total_tokens)
```
